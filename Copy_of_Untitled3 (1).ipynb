Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.


```python
import pandas as pd
import matplotlib.pyplot as plt


flight_prices= train_df['Price']

plt.figure(figsize=(10, 6))
plt.hist(flight_prices, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

##### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs'].unique()
```

```python
numeric_installs = pd.to_numeric(df['Installs'])
```

```python
install_counts = df['App'].value_counts().reset_index()
install_counts.columns = ['App', 'Installs']

# Sort the table by the number of installs in descending order
sorted_installs = install_counts.sort_values(by='Installs', ascending=False)

# Display the top 10 most popular apps
top_10_apps = sorted_installs.head(10)
print("Top 10 Most Popular Apps:")
print(top_10_apps)
```
